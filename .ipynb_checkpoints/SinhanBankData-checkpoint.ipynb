## 연도별 추이 분석하기 

```python
total_data['기준년월'].unique()
```

```python
기준년월별_예적금금액변동 = pd.DataFrame(total_data.groupby(['기준년월','지역구'])['예적금금액'].sum())
```

```python
기준년월별_예적금금액변동 = 기준년월별_예적금금액변동.reset_index()
기준년월별_예적금금액변동
```

```python
기준년월별_예적금금액변동["기준년월"] = pd.to_datetime(기준년월별_예적금금액변동["기준년월"],format="%Y%m")
```

```python
기준년월별_예적금금액변동["기준년월"].unique()
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns 

# 한글폰트 설정 
plt.rcParams['font.family'] = 'Malgun Gothic'
# 마이너스 폰트 깨짐 방지 
plt.rcParams['axes.unicode_minus'] = False
```

```python
plt.figure(figsize=(15,15))

sns.lineplot(data = 기준년월별_예적금금액변동, x='기준년월',y='예적금금액',hue='지역구')
plt.xticks(['2019-03-01T00:00:00.000000000', '2019-09-01T00:00:00.000000000',
       '2020-03-01T00:00:00.000000000', '2020-09-01T00:00:00.000000000',
       '2021-03-01T00:00:00.000000000'],['201903','201909','202003','202009','202103'])
```

전체적으로 예적금 금액의 변동이 20년 3월을 기점으로 변동함을 볼 수 있습니다.  
다만, 지역별로 금액의 차이가 커, 나누어 살펴보는것이 좋을 것 같습니다. 

```python
기준년월별_예적금금액변동['예적금금액'].describe()
```

```python
fig, (ax1,ax2,ax3,ax4) = plt.subplots(nrows=4, figsize=(50,50))

sns.lineplot(data = 기준년월별_예적금금액변동.loc[기준년월별_예적금금액변동['예적금금액']< 2.208864e+10], x='기준년월',y='예적금금액',hue='지역구', ax=ax1)
sns.lineplot(data = 기준년월별_예적금금액변동.loc[(기준년월별_예적금금액변동['예적금금액']> 2.208864e+10)  & (기준년월별_예적금금액변동['예적금금액'] < 2.637882e+10)], x='기준년월',y='예적금금액',hue='지역구', ax=ax2)
sns.lineplot(data = 기준년월별_예적금금액변동.loc[(기준년월별_예적금금액변동['예적금금액']> 2.637882e+10) & (기준년월별_예적금금액변동['예적금금액'] < 3.388205e+10)], x='기준년월',y='예적금금액',hue='지역구', ax=ax3)
sns.lineplot(data = 기준년월별_예적금금액변동.loc[기준년월별_예적금금액변동['예적금금액']> 3.388205e+10], x='기준년월',y='예적금금액',hue='지역구',ax=ax4)

# fig.xticks(['2019-03-01T00:00:00.000000000', '2019-09-01T00:00:00.000000000',
#        '2020-03-01T00:00:00.000000000', '2020-09-01T00:00:00.000000000',
#        '2021-03-01T00:00:00.000000000'],['201903','201909','202003','202009','202103'])



ax1.set_title("0~25%")
ax2.set_title("25~50%")
ax3.set_title("50~75%")
ax4.set_title("75~100%")
```
