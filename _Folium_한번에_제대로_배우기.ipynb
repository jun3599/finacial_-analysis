### PolyLineTextPath

```python
m = folium.Map([37,127], zoom_start=5)

wind_position = [[30.78443192761499, 125.38945157382388],
                [37.11318028062158, 128.31019000267548],
                [40.46586977021269, 132.25674871650887]]

wind_line = folium.PolyLine(
        wind_position,
        weight=20,
        color= 'deepskyble'
).add_to(m)

plugins.PolyLineTextPath(
    wind_line,
    ')',
    repeat = True,
    offset=7,
    attributes= {'fill':'dodgerblue', 'font-weight':'bold', 'font-size':'24'}
).add_to(m)

m
```

##### 그림 삽입하기 

```python
m = folium.Map(location=[37.552268, 126.991248], zoom_start=11)

# step 1. 그림 객체를 생성합니다. 
figure = branca.element.Figure()
# step 2. 그림 객체에 그림을 입력합니다. (여기에서는 간단히 더 확대된 버전의 지도사진을 포함시킵니다.)
folium.Map(location=[37.552268, 126.991248], zoom_start=15).add_to(figure)

# step 3. 그림 객체가 들어갈 프레임 구조를 생성합니다. 
iframe = branca.element.IFrame(width=500, height=300)
# step 4. 생성한 프레임 위치에 기존에 생성한 그림객체를 삽입합니다. 
figure.add_to(iframe)

# step 5. 팝업창 객체 안에 위에서 생성한 프레임 객체를 넣어줍니다. 
popup = folium.Popup(iframe, max_width=2650)

# step6. 마커를 생성하고 해당 마커 안에 데이터로 넣어줍니다. 
folium.Marker(
    [37.552268, 126.991248],
    popup=popup
).add_to(m)

m
```

### 마커 클러스터(Marker Cluster)
* 마커를 여러개 생성하여 운용합니다. 
* 그림을 축소할수록 주변의 마커들이 하나의 단위로 통합됩니다. 
* 확대 될수록 군집이 풀리며, 최소단위는 단일 마커입니다. 

```python
# 마커의 숫자 지정 
N = 100 
# N쌍의 랜덤한 위치값 생성 (임의지정, 추후에는 원하는 좌표들을 추가할 수 있음)
data = np.array([np.random.uniform(low=35.5, high=37.5, size=N),
                np.random.uniform(low=127, high=129, size=N)]).T

# 데이터 갯수 만큼의 popup을 생성합니다. (마찬가지로 내용 자체는 원하는 것을 넣을 수 있습니다.)
popups = [str(i) for i in range(N)]

m = folium.Map([36.5, 128], zoom_start=8)
plugins.MarkerCluster(data, popups=popups).add_to(m)

m
```

```python
import vincent

# vincent를 통해 산포도를 삽입합니다. #############################

# step1. 랜덤값으로 데이터 생성(추후 자료를 삽입할 수 있습니다.)  #
scatter_points = {
    'x': np.random.randn(50).cumsum(),
    'y': np.random.randn(50).cumsum()
}

# step2. vincent를 활용해 html 형식의 차트를 생성합니다. 
scatter_chart = vincent.Scatter(scatter_points,
                               iter_idx= 'x',
                               width= 400,
                               height= 200)

# step3. 차트를 json 코드로 변환합니다. 
scatter_json = scatter_chart.to_json()
# step4. dictionary 타입으로 변환 
scatter_dict = json.loads(scatter_json)
######################################################

# 폴리움 객체에 추가해줍니다. 
m = folium.Map([36.5, 128], zoom_start=7)

popup = folium.Popup()
folium.Vega(scatter_chart, height=250, width=450).add_to(popup)
folium.Marker([36, 128], popup=popup).add_to(m)

popup = folium.Popup()
folium.Vega(scatter_json, height=250, width=450).add_to(popup)
folium.Marker([37, 128], popup=popup).add_to(m)

popup = folium.Popup()
folium.Vega(scatter_dict, height=250, width=450).add_to(popup)
folium.Marker([38, 128], popup=popup).add_to(m)

m
```

```python
# 예제가 있는 url에서 직접 그림을 가져와 시각화를 진행합니다. 
url = "https://raw.githubusercontent.com/python-visualization/folium/master/examples/data"
vis1 = json.loads(requests.get(f'{url}/vis1.json').text)
vis2 = json.loads(requests.get(f'{url}/vis2.json').text)
vis3 = json.loads(requests.get(f'{url}/vis3.json').text)

m = folium.Map(
        location = [47.3489, -124.708],
        zoom_start=7,
        titles = 'Stamen Terrain'
)

folium.Marker(
    [47.3489, -124.708],
    popup= folium.Popup(max_width=450).add_child(
                                folium.Vega(vis1, width=450, height=250))).add_to(m)
    
folium.Marker(
    [44.639, -124.5339],
    popup= folium.Popup(max_width=450).add_child(
                                folium.Vega(vis2, width=450, height=250))).add_to(m)
folium.Marker(
    [46.216, -124.1280],
    popup= folium.Popup(max_width=450).add_child(
                                folium.Vega(vis3, width=450, height=250))).add_to(m)    

m
```

```python
# 시간에 따라 그림이 변화하는 지도를 그립니다. 

m = folium.Map([37.521628, 126.9241247], zoom_start=15)

lines = [
    {'coordinates': [[126.924127, 37.521628],
                   [126.921754,37.524117]],
    'dates': ['2020-01-01T00:00:00', '2020-01-01T00:10:00'],
    'color': 'red', 'weight':10
    },
    
    {'coordinates': [[126.921754,37.524117],
                   [126.926168,37.526768]],
    'dates': ['2020-01-01T00:10:00', '2020-01-01T00:20:00'],
    'color': 'yellow'
    },
    {'coordinates': [[126.926168,37.526768],
                   [126.928490, 37.524340]],
    'dates': ['2020-01-01T00:20:00', '2020-01-01T00:30:00'],
    'color': 'green'
    },
    {'coordinates': [[126.928490,37.524340],
                   [126.924127, 37.521628]],
    'dates': ['2020-01-01T00:30:00', '2020-01-01T00:40:00'],
    'color': 'blue'
    }
    
]
# 스타일 지정 
features = [
    {'type': 'Feature',
    'geometry':{
        'type': 'LineString',
        'coordinates': line['coordinates'],},
     
    'properties':{
        'times': line['dates'],
        'style':{
            'color': line['color'], 
            'weight': line['weight'] if 'weight' in line else 5 }}
    }for line in lines 
]

plugins.TimestampedGeoJson({
    'type': 'FeatureCollection',
    'features': features,
},period= 'PT1M', add_last_point=True).add_to(m)


m
# cordinate 좌표 - dates 시간 쌍 입력 
```
