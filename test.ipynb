```python
distortions = []
for i in range(1, 50):
    km = KMeans(n_clusters=i, 
                init='k-means++', 
                n_init=10,
                max_iter= 512, 
                random_state=87)
    km.fit(data)
    
    distortions.append(km.inertia_)
    
    
plt.plot(range(1, 50), distortions, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
plt.tight_layout()
plt.show()
```

```python
from sklearn import metrics
from sklearn.metrics import silhouette_score
from numpy import unique

distortions = []
score = []
for i in range(2, 50):
    km = KMeans(n_clusters=i, 
                init='k-means++', 
                n_init=16,
                max_iter= 512, 
                random_state=807)
   
    y_hat = km.fit_predict(data)
    
    clusters_km = unique(y_hat)
    labels_km = km.labels_
    score_km = metrics.silhouette_score(data,labels_km)
    
    distortions.append(km.inertia_)
    score.append(score_km)
    
# plt.plot(range(2, 50), distortions, marker='o')
# # plt.plot(range(2, 50), score, marker='o')
# plt.xlabel('Number of clusters')
# plt.ylabel('Distortion')
# plt.tight_layout()
# plt.show()



fig, ax1 = plt.subplots()
ax1.plot(range(2, 50), distortions, marker='o')
ax2 = ax1.twinx()
ax2.plot(range(2, 50), score, marker='x',color='red')

plt.show()
```

```python
from sklearn.cluster import SpectralClustering

score = []
for i in range(2,30):
    sp = SpectralClustering(n_clusters=i,
                eigen_solver = 'arpack',
                affinity = 'nearest_neighbors',
                n_init=10,  
                random_state=1)
    y_hat = sp.fit_predict(data)
    
    clusters_sp = unique(y_hat)
    labels_sp = sp.labels_
    score_sp = metrics.silhouette_score(data,labels_sp)
    
    score.append(score_sp)
    
plt.plot(range(2, 30), score, marker='o', color='red')
plt.xlabel('Number of clusters')
plt.ylabel('score')
plt.tight_layout()
plt.show()
```
